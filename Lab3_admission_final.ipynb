```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/ML_Assignments/datasets/admission.csv")
```

```python
#1. Scatter Plot â€“ GRE Score vs. Chance of Admit

plt.figure(figsize=(7, 5))
sns.scatterplot(data=df, x='GRE Score', y='Chance of Admit ')
plt.title('Scatter Plot: GRE Score vs Chance of Admit')
plt.xlabel('GRE Score')
plt.ylabel('Chance of Admit')
plt.show()
```

```python
#2. 2. Bar Plot â€“ Average CGPA for Each Research Status

plt.figure(figsize=(7, 5))
sns.barplot(x='Research', y='CGPA', data=df)
plt.title('Bar Plot: Avg CGPA vs Research Status')
plt.xlabel('Research (0 = No, 1 = Yes)')
plt.ylabel('Average CGPA')
plt.show()
```

```python
#3. Box Plot â€“ Distribution of TOEFL Score

plt.figure(figsize=(6, 5))
sns.boxplot(data=df['TOEFL Score'], color='lightgreen')
plt.title('Box Plot: TOEFL Score')
plt.xlabel('TOEFL Score')
plt.show()
```

```python
#4. Pie Chart â€“ Proportion of Students with/without Research Experience

plt.figure(figsize=(6, 6))
df['Research'].value_counts().plot.pie(autopct='%1.1f%%', labels=['Yes', 'No'], colors=['skyblue', 'pink'])
plt.title('Pie Chart: Research Experience')
plt.ylabel('')
plt.show()
```

```python
# Sort the dataset by GRE Score
df_sorted = df.sort_values(by="GRE Score")

# Create a line chart: GRE Score vs. Chance of Admit
plt.figure(figsize=(10, 6))
plt.plot(df_sorted["GRE Score"], df_sorted["Chance of Admit "], marker='o', linestyle='-')
plt.title("GRE Score vs Chance of Admission")
plt.xlabel("GRE Score")
plt.ylabel("Chance of Admission")
plt.grid(True)
plt.tight_layout()
plt.show()
```
