```python
import numpy as np
import pandas as pd
```

```python
#Reading data from csv
df = pd.read_csv('HeartData.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='MaxHR', hue='ChestPain', data=df)
plt.title('Age vs MaxHR')
plt.xlabel('Age')
plt.ylabel('MaxHR')
plt.show()
```

```python
#Bar Plot
plt.figure(figsize=(8, 6))
df['ChestPain'].value_counts().plot(kind='bar')
plt.title('Chest pain type vs Frequency')
plt.xlabel('ChestPain Type')
plt.ylabel('Frequency')
plt.show()
```

```python
#Box Plot
plt.figure(figsize=(8, 6))
sns.boxplot(y='Age', data=df)
plt.title('Box Plot for Age')
plt.ylabel('Age')
plt.show()
```

```python
#Pie Chart
plt.figure(figsize=(8, 6))
plt.pie(df['Thal'].value_counts(), labels=df['Thal'].value_counts().index, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.ylabel('')
plt.show()
```

```python
#Line Chart
AgeArr = np.array(df['Age'])
SortedAgeArr = np.sort(AgeArr)
plt.figure(figsize=(24, 6))
plt.plot(SortedAgeArr, df['Chol'], marker='.')
plt.title('Age vs Chol')
plt.xlabel('Age')
plt.ylabel('Cholesterol (Chol)')
plt.grid(True)
plt.show()
```
